```python
import numpy as np
import matplotlib.pyplot as plt
```

Experimento 1

```python
f = 10000
A = 1
T = 1/f


tempo = np.linspace(0,3*T,1000)
y = A*np.sin(2*np.pi*f*tempo)

plt.plot(tempo, y)

plt.xlabel("Tempo em Segundos(s)")
plt.ylabel("Amplitude (V)")
plt.title("Grfico Exerccio 1 - Onda Senoidal de 10kHz")
plt.grid(True)
plt.figure(1)
plt.show()
```

```python
f = 10000
A = 1
T = 1/f


tempo = np.linspace(0,3*T,100)
y = A*np.sin(2*np.pi*f*tempo)

print("Tabela de Amostras ")
print("   N | Tempo (s)  | Amplitude (V)")
for i in range(len(tempo)):
  print(f" {i+1:3d} | {tempo[i]:.8f} | {y[i]:.8f}")

  


plt.plot(tempo, y)
plt.xlabel("Tempo em Segundos(s)")
plt.ylabel("Amplitude (V)")
plt.title("Grfico Exerccio 2 - Onda Senoidal de 10kHz")
plt.grid(True)
plt.figure(2)
plt.show()

```
